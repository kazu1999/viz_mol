テキストファイル内にあるSMILESを可視化するツール

```python
from rdkit import Chem
```

```python
# ファイル読み込み
path = 'pubchem_1159604.txt'
with open(path) as f:
    smiles = f.read().splitlines()
print(smiles)
```

```python
mols = []
for smi in smiles:
    m = Chem.MolFromSmiles(smi)
    mols.append(m)
Chem.Draw.MolsToGridImage(mols, molsPerRow=4, subImgSize=(200,200))
```
